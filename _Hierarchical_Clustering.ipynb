```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics.pairwise import euclidean_distances
```

```python
X = np.array([[0.40, 0.53],
              [0.22, 0.38],
              [0.35, 0.32],
              [0.26, 0.19],
              [0.08, 0.41],
              [0.45, 0.30]])
```

```python
X
```

```python
distance_matrix = euclidean_distances(X, X)
```

```python
distance_matrix
```

```python
Z = linkage(X, method='sinle', metric='euclidean')
```

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")
dendrogram(Z, labels=['P1', 'P2', 'P3', 'P4', 'P5', 'P6'])
plt.show()
```
